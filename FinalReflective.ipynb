```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
dataset=pd.read_csv("dataset_urop.csv")
```

```python
df_copy = dataset.copy()
df_copy
```

```python
df_copy = pd.DataFrame(dataset)

# Drop the last 4 columns
df_copy = df_copy.iloc[:, :-4]
df_copy
```

```python
import pandas as pd

# Assuming your original DataFrame is named 'df'
# Replace this with your actual DataFrame



df = pd.DataFrame(dataset)

# Create a new DataFrame with the split columns
split_df = pd.DataFrame()

# Split the last 4 columns into 8 columns
for col in df.columns[-4:]:
    split_df[col.split('_')[0]] =abs( df[col] - (-11))
    split_df[col.split('_')[1]] = abs(df[col] - 11)

# Set the index to be 0, 1, 2, ..., 13
split_df.index = range(len(split_df))

# Print the resulting DataFrame
print(split_df)
```

```python
df1 = pd.DataFrame(df_copy)
df2 = pd.DataFrame(split_df)

new_dataset = pd.concat([df1, df2], axis=1)


new_dataset
```

```python
last_8_columns = new_dataset.iloc[:, -8:]

# Finding the minimum value in each row of the last 8 columns
min_values = last_8_columns.min(axis=1)

# Creating a new DataFrame with 1's where minimum and 0's otherwise
new_data = np.where(last_8_columns.eq(min_values, axis=0), 1, 0)

# Appending the new data to the existing DataFrame
new_columns = [f"{col}" for col in last_8_columns.columns]  # Creating new column names
new_df = pd.concat([df_copy, pd.DataFrame(new_data, columns=new_columns)], axis=1)

new_df
```

```python
import pandas as pd

# Assuming df is your DataFrame
# List of learning styles columns
learning_styles = ['active', 'reflective', 'sensing', 'intuitive', 'visual', 'verbal', 'sequential', 'global']

# Create a new column 'learning_style' with the index of the maximum value column
new_df['learning_style'] = new_df[learning_styles].idxmax(axis=1).apply(lambda x: learning_styles.index(x) + 1)

# Display the DataFrame
new_df
```

```python
new_df = new_df.drop(['active', 'sensing', 'intuitive', 'visual', 'verbal', 'sequential', 'global'], axis=1)
```
