```python
import numpy as np

# Assuming you have matrices named confidence_matrix1, confidence_matrix2, ..., confidence_matrix8

# Combine matrices into a list for easy iteration
matrices = [confidence_matrix1, confidence_matrix2, confidence_matrix3, confidence_matrix4,
            confidence_matrix5, confidence_matrix6, confidence_matrix7, confidence_matrix8]

# Total number of columns (assuming 1 column in each matrix)
total_columns = len(matrices)
print(total_columns)

# Create an empty array to store the results for each row
result_matrix = np.zeros((200, 2))

# Iterate over rows and find the maximum value and its index for each row
for row in range(200):
    max_value, max_column_index = max((matrix[row], matrix_index + 1)
                                      for matrix_index, matrix in enumerate(matrices))
    
    # Store the results in the result_matrix
    result_matrix[row, 0] = max_value
    result_matrix[row, 1] = max_column_index

# Print or use result_matrix as needed
print(result_matrix)
```
